### INFORME TÉCNICO

#### 1. Introducción

##### Se describe el desarrollo y funcionamiento de una calculadora básisca implementada en el lenguaje de programación Python aplicando de manera correcta los pricipios de POO, como son: herencia , encapsulación y polimorfismo. El sistema se estructura en diferentes modulos para la mejor organización, reutilización y mantenimiento del código.

#### 2. Objetivo

##### Desarrollar una calculadora básica, con la capacidad de realizar operaciones fundamentales: Suma, resta, multiplicación y división

#### 3. Arquitectura del proyecto

##### Modelos/operación: clase base abstracta. Modelos/operación: Clase derivada que implementa las operaciones. Servicios/calculadora_servicio.Py: Gestiona la lógica del cálculo. main.py: punto de entrada del programa y ejecución del mismo.

#### 4. Descripción de clases


#### 4.1 Clase operación (Clase base)

##### La clase operaciónn representa la clase padre del programa, en ella se aplican los principios de herencia y encapsulación, contiene los atributos protegidos _número1 y _número2, también define el método calcular() como un método genérico que será sobreescrito por las clases hijas estableciendo una estructura común

#### 4.2 Clase OperaciónBásica (Clase derivada)

##### La clase OperaciónBásica hereda de Operación y aplica el pricipio de polimorfismo, ya que sobreescribe el método calcular(). Según el operador ingresado por el usuario el método realiza: Suma(+), Resta(-), Multiplicación(*) y División(/),validando que el divisor no sea 0

#### 4.3 Clase CalculadoraServicio (Clase de servicio)

##### Contiene la lógica pricipal del sistema, su función es: Crear unas instancia de OperacionBásica, enviar los datos enviados por el usuario, ejecuatar el cálculo correspondiente, diseño que permite  separar la lógica del resto del programa.

#### 5. Funcionamiento del programa (main.py)

##### Actua como punto de netrada del sistema: Muestra un mensaje inicial, solicita al usuario el ingreso de dos números, solicita el operador aritmetico, llama al servicio CalculadoraServicio ejecuta y muetra el resultado final en la pantalla.

##### 6. Principios de Programación Orientada a Objetos aplicados

##### Encapsulación: uso de atributos protegidos para controlar el acceso a los datos. Herencia: OperaciónBásica hereda atributos y métodos de Operación. Polimorfismo:El método calcualar() es redefinido según la necesidad de la clase derivada

#### 7. Conclusión

##### La calculadora básica desarrollada demuestra una correcta aplicación de la POO en Python. La separación por capas, el uso de herencia, polimorfismo y la organización modualar del código permiten que el sistema sea fácil de entender, mantener y ampliar, por ejemplo incoorporando nuevas operaciones matemáticas y interfases gráficas en el futuro. Este programa constituye una base sólida para el aprendizaje y la aplicación práctica de los conceptos fundamentales de POO.
