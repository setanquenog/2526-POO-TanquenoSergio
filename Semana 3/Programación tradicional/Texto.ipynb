## INTRODUCCÓN

#### La programación es un proceso, que crea y ejecuta instrucciones que permiten resolver problemas, existiendo múltiples paradigmas de programación, entre los cuales los más utillizados se encuentran la programación trádicional y la programación orientada a objetos POO, donde cada una tiene cracterísticas particulares que influyen en la forma de estructurar y organizar el código.

### PROGRAMACIÓN TRADICIONAL

#### Se denomina tambien procedimental, se centra en el uso de funciones que se ejecutan de manera secuencial, donde los datos y operaciones se manejan por separado, el programa sigue un flujo logico paso a paso. Se menciona en el ejemplo de promedio semanal de temperatura, donde se emplea funciones independientes para el ingreso de datos y poder calcular el promedio, las temoeraturas se almacenan en una lista y son manipuladas directamente por las funciones, sin referir al clima como una entidad especifíca.

### PROGRAMACIÓN ORIENTADA A OBJETOS (POO)

#### Se fundamenta principalmente en la organización del software mediante clases y objetos que represnetan conceptos del mundo real puediendo destacar que este paradigma se basa en los principios de abstracción, encapsulamiento, herencia y polimorfismo. Se menciona como ejemplo el promedio semanal de temperaturas donde la POO, ejecuta las siguientes clases. CLASE CLIMA: representa la información general del clima; CLASE CLIMA SEMANAL:gestión de los datos climáticos de cada día de la semana; CLASE APLICACIÓN CLIMA: control y ejecución del programa. Donde los datos y los metodos se integran en una sola estructura, generando un código más organizado y reutilizable.

### CONCLUSIÓN

#### Se puede determinar que la Programación Tradicioanl se adecua para programas simples en razón de su estructura directa y comprensión fácil, se debe mencionar que presenta limitaciones en programs más complejos. En cuando la Programación Orientada a Objetos ejecuta una mejor organización del código, facilita el mantenimiento y promueve la reutilización del código, se debe reslatar en el ejemplo analizado, que la programación orientada a obejetos genera una solución más eficiente y estructurada en el manejo de la información climática.
